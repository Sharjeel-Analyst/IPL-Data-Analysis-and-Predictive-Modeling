# üèè IPL Dataset Analysis with Python

```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Load data
deliveries_df = pd.read_csv("deliveries.csv")
matches_df = pd.read_csv("matches.csv")
```

```python
# Show head and nulls
print("Deliveries Head:")
deliveries_df.head()
```

```python


print("\nMatches Head:")
matches_df.head()
```

```python

print("\nDeliveries Nulls:")
deliveries_df.isnull().sum()
```

```python

print("\nMatches Nulls:")
matches_df.isnull().sum()
```

```python
# Descriptive statistics
print("Deliveries Description:")
deliveries_df.describe(include='all')
```

```python

print("\nMatches Description:")
matches_df.describe(include='all')
```

```python
# display all columns
pd.set_option('display.max_columns', None)
```

```python
# Merge data
merged_df = deliveries_df.merge(matches_df, how='left', left_on='match_id', right_on='id')
print("Merged Data Head:")
merged_df.head()
```

```python
# Top Batters and Bowlers
top_batters = merged_df.groupby("batter")["batsman_runs"].sum().sort_values(ascending=False).head(10).reset_index()
print("Top Run Scorers:")
top_batters
```

```python
wickets = merged_df[merged_df["is_wicket"] == 1]
top_bowlers = wickets.groupby("bowler").size().sort_values(ascending=False).head(10).reset_index(name='wickets')
print("Top Wicket Takers:")
top_bowlers
```

```python

matches_per_season = matches_df["season"].value_counts().sort_index().reset_index()
matches_per_season.columns = ["season", "matches"]
print("Matches per Season:")
matches_per_season
```

```python
# Plot Top Batters
plt.figure(figsize=(10, 6))
plt.bar(top_batters["batter"], top_batters["batsman_runs"])
plt.title("Top 10 Run Scorers in IPL")
plt.xlabel("Batter")
plt.ylabel("Total Runs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### ‚úÖ Summary

- The analysis covers delivery-level and match-level IPL data.
- We identified top performers using groupby operations.
- Visualized top 10 run scorers.
- You can extend this analysis with strike rates, win margins, toss effects, etc.


## üìä Advanced EDA on IPL Dataset

```python
# Strike Rate = (Total Runs / Total Balls Faced) * 100
batter_stats = merged_df.groupby('batter').agg({
    'batsman_runs': 'sum',
    'ball': 'count'
}).rename(columns={'ball': 'balls_faced'}).reset_index()
batter_stats['strike_rate'] = (batter_stats['batsman_runs'] / batter_stats['balls_faced']) * 100
top_strike_rate = batter_stats[batter_stats['balls_faced'] > 200].sort_values(by='strike_rate', ascending=False).head(10)
print(top_strike_rate)
```

```python
dismissal_counts = merged_df[merged_df['is_wicket'] == 1]['dismissal_kind'].value_counts()
print("Wicket Types:")
print(dismissal_counts)
```

```python
team_wins = matches_df['winner'].value_counts().head(10)
print("Top Teams by Wins:")
print(team_wins)
```

```python
venue_counts = matches_df['venue'].value_counts().head(10)
print("Most Common Venues:")
print(venue_counts)
```

```python
venue_counts.head(5).plot(kind='bar', title='Top 5 Most Used Venues', xlabel='Venue', ylabel='Matches Hosted')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
